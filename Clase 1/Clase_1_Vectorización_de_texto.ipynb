### **Parte A) Tokenization**

#### **Ejercicio I: Crear un Tokenizador Básico**

Descripción: Escribe una función que tome una cadena de texto y la divida en palabras individuales (tokens)

```python
def tokenizar(texto):
    texto = texto.lower()

    signos_puntuacion = ['.', ',', '!', '?', ':', ';', '(', ')', '[', ']', '{', '}', '¿', '¡', '"', "'"]

    for signo in signos_puntuacion:
        texto = texto.replace(signo, ' ')

    tokens = texto.split()

    return tokens

texto = "¡Hola, Mundo! Este es un ejemplo de tokenización"
tokens = tokenizar(texto)
print(tokens)
```

#### **Ejercicio II: Crear un Tokenizador que maneje palabras compuestas**

Descripción: Escribe una función que tome una cadena de texto y la divida en palabras individuales (tokens) pero que no divida palabras compuestas, es decir que la palabra "New York" se mantenga como una sola palabra y no como "New", "York".

Ejemplo:
Entrada: "Mañana viajare a New York"
Salida: "mañana", "viajare", "a", "New York"

```python
def tokenizer(sentence, compound_words):

    sentence = sentence.lower()
    tokens = []

    for compound in compound_words:
        sentence = sentence.replace(compound.lower(), compound.replace(" ", "_"))

    words = sentence.split()

    for word in words:
        if "_" in word:
            tokens.append(word.replace("_", " "))
        else:
            tokens.append(word)

    return tokens


sentence = "Mañana viajare a New York"
compound_words = ["New York", "Los Angeles"]
tokens = tokenizer(sentence, compound_words)
print(tokens)
```

#### **Ejercicio V, Lemmatization con WordNetLemmatizer de NLTK**

Descripción: Utiliza NLTK para realizar lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

WordNet es una base de datos léxica grande del inglés. Se usa para encontrar la forma base de una palabra.

```python
nltk.download("wordnet")
#Es un recurso que proporciona WordNet en múltiples idiomas, ampliando así las capacidades de WordNet.
nltk.download("omw-1.4")
```

Inicializamos el lemmatizer

```python
lemm = WordNetLemmatizer()
```

Definimos ejemplos de palabras a aplicar el lemmatizer

```python
example_words = ["relate", "related", "relating", "relation", "relational"]
```

Aplicamos lemmatize de WordNetLemmatizer, especificando que la POS sea Verbo.

```python
for word in example_words:
  '''
  pos="v": Indica la parte del discurso (part-of-speech) de la palabra a lematizar. La lematización puede variar dependiendo del POS tag asignado a la palabra:

    pos="v": Verbo. Por ejemplo, "programming" se convierte en "program".
    Otros ejemplos:
        pos="n": Sustantivo.
        pos="a": Adjetivo.
        pos="r": Adverbio.
  Especificar el POS tag ayuda al lematizador a comprender el contexto y lematizar correctamente. Por defecto, el lematizador asume que las palabras son sustantivos si no se especifica el POS.
  '''

  print(lemm.lemmatize(word, pos="v"))
```

### **Parte C) Vectorización de Texto**

```python
documents = [
    "me gusto mucho la pelicula",
    "me gusto la pelicula",
    "la pelicula fue terrible",
    "odio realmente odio esta pelicula"
]
```


#### **Ejercicio V, Bag of Words (CountVectorizer)**

Importamos lo necesario

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
```

Utilizaremos CountVectorizer

```python
bow_vectorizer = CountVectorizer()
bow_matrix = bow_vectorizer.fit_transform(documents)

print("=== Bag of Words Matrix ===")
print(pd.DataFrame(bow_matrix.toarray(), columns=bow_vectorizer.get_feature_names_out()))
```

Aplicamos Similitud Coseno

```python
print("\nCosine Similarity (BOW):")
print(pd.DataFrame(cosine_similarity(bow_matrix)))
```

#### **Ejercicio VI, Binary Bag of Words**

Utilizaremos CountVectorizer aplicando el modo binario

```python
onehot_vectorizer = CountVectorizer(binary=True)
onehot_matrix = onehot_vectorizer.fit_transform(documents)
```

```python
print("\n=== One Hot Encoding Matrix ===")
print(pd.DataFrame(onehot_matrix.toarray(), columns=onehot_vectorizer.get_feature_names_out()))
```

Aplicamos Similitud Coseno

```python
print("\nCosine Similarity (One-Hot):")
print(pd.DataFrame(cosine_similarity(onehot_matrix)))
```

#### **Ejercicio VII, TF-IDF**

Utilizaremos TfidfVectorizer

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

print("\n=== TF-IDF Matrix ===")
print(pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out()))
```

Aplicamos Similitud Coseno

```python
print("\nCosine Similarity (TF-IDF):")
print(pd.DataFrame(cosine_similarity(tfidf_matrix)))
```

#### **Ejercicio VIII, Distancia Euclideana vs Similitud Coseno**

**Parte A**

En el primer caso definiremos dos "embeddings"

```python
v1 = np.array([1, 2, 3])
v2 = np.array([2, 4, 6])
```

Calculamos la Distancia Euclideana

```python
euclidean_distance = np.linalg.norm(v1 - v2)
print("Distancia Euclidiana:", euclidean_distance)
```

Calculamos la Similitud Coseno

```python
dot_product = np.dot(v1, v2)
norm_v1 = np.linalg.norm(v1)
norm_v2 = np.linalg.norm(v2)

cosine_similarity = dot_product / (norm_v1 * norm_v2)
print("Similitud Coseno:", cosine_similarity)
```

### **Parte D) Machine Learning para NLP**

Importamos lo necesario

```python
from sklearn.naive_bayes import MultinomialNB
```

Definimos (nuevamente) los documentos

```python
docs = [
    "me gustó la película",
    "excelente película",
    "mala película",
    "película horrible"
]
```

Definimos las etiquetas

```python
y = ["pos", "pos", "neg", "neg"]
```

Aplicamos Bag of Words

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(docs)
```

Aplicamos Naive Bayes Multinomial

```python
model = MultinomialNB()
model.fit(X, y)
```

Realizamos la predicción de una nueva oración

```python
nuevo = vectorizer.transform(["excelente película"])
pred = model.predict(nuevo)

print("Predicción:", pred[0])
```
